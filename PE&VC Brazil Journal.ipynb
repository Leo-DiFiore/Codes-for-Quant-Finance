```python
from bs4 import BeautifulSoup as BS
import requests
from collections import Counter

url = "https://braziljournal.com/hot-topic/private-equity-vc"
html = requests.get(url).text
soup = BS(html, "html5lib")

set_inicial = {linha["href"] for linha in soup("a") if linha.has_attr("href") and Counter(linha.text)[" "] >= 3}
noticias = [linha for linha in set_inicial if linha.startswith("https://braziljournal.com/") and not "hot-topic" in linha and not "/tag/" in linha]

noticias
```
